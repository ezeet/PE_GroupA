### Lottery Design
***

1.	 Participants have to give responses in 30 trials

2.	 Within-subject condition is the amount the lottery pays out

3.	 Between-subject condition is the difficulty 
	(diffierence in probability between two gambles, safe vs. risky)

4.	 Randomization of left vs. right display of the gamble variants

```python
from __future__ import division
from psychopy.visual import ImageStim, TextStim, Window
from psychopy import core, event, gui, data, logging
import numpy as np
import pandas as pd
import os

from routines import Routine

#general settings
expName = 'Group A'
screen_size = [2000, 1200]
frames_per_second = 60
full_screen = False
background_color = '#eeeeee'
n_trials = 30

# trial settings
choice_keys = ['q', 'p']
escape_key = 'escape'
choice_time_limit = 5
feedback_duration = 2
fixation_duration = 1

#stimuli settings
text_color = 'blue'
text_height = 50
options_x_offset = 400
image_size = 100

#store info about the experiment session
dlg = gui.Dlg(title=expName)
dlg.addField('Participant:', 1)
dlg.addField('Age:', 25)
dlg.addField('Gender:', choices=['female', 'male', 'prefer not to disclose'])
dlg.addField("condition", choices = ["A", "B"]) #change into A and B
dlg.addField('Handedness:', choices=['right', 'left', 'both'])
dlg.show()

expInfo = dict(zip(['participant', 'age', 'gender', 'condition', "hand"], dlg.data))
expInfo['date'] = data.getDateStr()  # add a simple timestamp
expInfo['expName'] = expName # add the experiment name
if dlg.OK:  # then the user pressed OK
    print(expInfo)
else:
    print(expInfo)
    core.quit()

#check if data folder exists
directory=os.path.join(os.getcwd(), 'data')
if not os.path.exists(directory):
    os.makedirs(directory)

#create file name for storing data
fileName = os.path.join('data', '%s_%s_%s' % (expName, expInfo['participant'], expInfo['date']))

#save a log file
logFile = logging.LogFile(fileName + '.log', level=logging.DEBUG)
logging.console.setLevel(logging.WARNING)  # this outputs to the screen, not a file

#create a window
mywin = Window(screen_size, units='pix', color=background_color, fullscr=full_screen)

#create some stimuli
left_feedback = TextStim(win=mywin, color=text_color, pos=(-options_x_offset, 0), height=text_height)
right_feedback = TextStim(win=mywin, color=text_color, pos=(options_x_offset, 0), height=text_height)

left_stimuli = TextStim(win=mywin, color=text_color, pos=(-options_x_offset, 0), height=text_height)
right_stimuli = TextStim(win=mywin, color=text_color, pos=(options_x_offset, 0), height=text_height)

#Lotterien (2x2 outcomes, ähnlichem EV, verschiedene Varianz)
#Zufällige Zahlen 1 bis 10 x 30
a = np.random.choice(a = np.arange(1,11), size = 30, replace = True)

#Zufällige Wahrscheinlichkeiten x 30 (px)
b = np.round(a = np.random.choice(a = np.arange(0.3, 0.7, 0.05), size = 30, replace = True), decimals = 2)
b2 = np.round(a = np.random.choice(a = np.arange(0.3, 0.7, 0.05), size = 30, replace = True), decimals = 2)
#EVx: x * px
c = a * b

# Y = evx/py
d = np.round(a = (c / (1 - b)), decimals = 2)
d2 = np.round(a = (c / (1 - b2)), decimals = 2)
#py: 1 - px
e = 1 - b
e2 = 1 - b2
#Evy: y * py
f = d * e


a2 = np.round(a = (c / b2), decimals = 2)
#creating data frame for nornal Lotteries
x = {"x": a, "px": b, "evx": c, "y": d, "py": e, "evy": f, "x2": a2, "px2": b2, "evx2": c, "y2": d2,  "py2": e2, "evy2": f}
normallot = pd.DataFrame(x)
normallot

#Lotterien (plus 1000 bei outcomes) #within subject
# Lotterie mit grossen Outcomes
# x + 1000
abig = np.random.choice(a = np.arange(1,11), size = 30, replace = True) + 1000
# px (gleich wie bei normal lotteries)
bbig = b
bbig2 = b2 
# evx = abig * bbig
cbig = abig * bbig
# Y = evx/py
dbig = np.round(a= (cbig / (1 - bbig)),decimals = 2) 
dbig2 = np.round(a = (cbig / (1 - bbig2)), decimals = 2)
# py = 1 - px
ebig = 1 - bbig
ebig2 = 1 - bbig2
# evy = y * py
fbig = dbig * ebig
#varx
gbig = {"x": abig, "px": b}
hbig = pd.DataFrame(gbig)
ibig = np.var(hbig, axis = 1)

abig2 = np.round(a = (cbig / bbig2), decimals = 2)
#vary
jbig = {"y": dbig, "py": ebig}
kbig = pd.DataFrame(jbig)
lbig = np.var(kbig, axis = 1)
#creating data frame
x = {"x": abig, "px": bbig, "evx": cbig, "y": dbig, "py": ebig, "evy": fbig, "x2": abig2, "px2": bbig2, "evx2": cbig, "y2": dbig2,  "py2": ebig2, "evy2": fbig}
biglot = pd.DataFrame(x)

#Lotterien mit rare event 
#Zufällige Zahlen 1 bis 10 x 30
ahard = np.random.choice(a = np.arange(1,11), size = 30, replace = True)

#Zufällige Wahrscheinlichkeiten x 30 (px zwieschen 0.1 und 0.3)
bhard = np.random.choice(a = np.arange(0.1, 0.3, 0.05), size = 30, replace = True)
bhard2 = np.random.choice(a = np.arange(0.1, 0.3, 0.05), size = 30, replace = True)
#EVx: x * px
chard = ahard * bhard

# Y = evx/py
dhard = np.round(a = (chard / (1 - bhard)), decimals = 2)
dhard2 = np.round(a = (chard / (1 - bhard2)), decimals = 2)
#py: 1 - px
ehard = 1 - bhard
ehard2 = 1 - bhard2
#Evy: y * py
fhard = dhard * ehard

#varx
ghard = {"x": ahard, "px": bhard}
hhard = pd.DataFrame(ghard)
ihard = np.var(hhard, axis = 1)

ahard2 = np.round(a = (chard / bhard2), decimals = 2)
#vary
jhard = {"y": dhard, "py": ehard}
khard = pd.DataFrame(jhard)
lhard = np.var(khard, axis = 1)

#creating data frame for nornal Lotteries
x = {"x": ahard, "px": bhard, "evx": chard, "y": dhard, "py": ehard, "evy": fhard, "x2": ahard2, "px2": bhard2, "evx2": chard, "y2": dhard2,  "py2": ehard2, "evy2": fhard}
hardlot = pd.DataFrame(x)

#Lotterien mit hohen outcomes und rare events
#Zufällige Zahlen 1 bis 10 x 30
abigh = abig

#Zufällige Wahrscheinlichkeiten x 30 (px zwieschen 0.1 und 0.3)
bbigh = bhard
bbigh2 = bhard2
#EVx: x * px
cbigh = abigh * bbigh

# Y = evx/py
dbigh = np.round(a = (cbigh / (1 - bbigh)), decimals = 2)
dbigh2 = np.round(a = (cbigh / (1 - bbigh2)), decimals = 2)
#py: 1 - px
ebigh = 1 - bbigh
ebigh2 = 1 - bbigh2
#Evy: y * py
fbigh = dbigh * ebigh

#varx
gbigh = {"x": abigh, "px": bbigh}
hbigh = pd.DataFrame(gbigh)
ibigh = np.var(hbigh, axis = 1)

abigh2 = np.round(a = (cbigh / bbigh2), decimals = 2)
#vary
jbigh = {"y": dbigh, "py": ebigh}
kbigh = pd.DataFrame(jbigh)
lbigh = np.var(kbigh, axis = 1)

#creating data frame for nornal Lotteries
x = {"x": abigh, "px": bbigh, "evx": cbigh, "y": dbigh, "py": ebigh, "evy": fbigh, "x2": abigh2, "px2": bbigh2, "evx2": cbigh, "y2": dbigh2,  "py2": ebigh2, "evy2": fbigh}
bighlot = pd.DataFrame(x)

#fixation cross
fixation_cross = TextStim(win=mywin, text='+', color=text_color)

#create the dataframe
data = pd.DataFrame([])

#draw the stimuli
trial_routine = Routine(window=mywin, frames_per_second=frames_per_second, escape_key=escape_key)

sequence = np.tile([0,1], 15)

for t in range(n_trials):
    # put here things that change every trial$
    if "condition" == "A": #A = easy option/condition
        stimuli = {"stimuli_sequence" :      [sequence],
                    'left_stimuli_normal':   normallot.loc[t, ["x", "px", "y", "py"]],
                    "left_stimuli_big" :     biglot.loc[t, ["x", "px", "y", "py"]],
                    'right_stimuli_normal':  normallot.loc[t, ["x2", "px2", "y2", "py2"]],
                    "right_stimuli_big":     biglot.loc[t, ["x2", "px2", "y2", "py2"]],
                    'left_feedback_normal':  np.random.choice(a = normallot.loc[t, ["x", "y"]],  p= normallot.loc[t, ["px", "py"]]),
                    "left_feedback_big" :    np.random.choice(a = biglot.loc[t, ["x", "y"]],  p= biglot.loc[t, ["px", "py"]]), 
                    'right_feedback_normal': np.random.choice(a = normallot.loc[t, ["x2", "y2"]],  p= normallot.loc[t, ["px2", "py2"]]), 
                    "right_feedback_big":    np.random.choice(a = biglot.loc[t, ["x2", "y2"]],  p= biglot.loc[t, ["px2", "py2"]])
                  }
    else:
         stimuli = {"stimuli_sequence" :     [sequence],
                    'left_stimuli_normal':   hardlot.loc[t, ["x", "px", "y", "py"]],
                    "left_stimuli_big" :     bighlot.loc[t, ["x", "px", "y", "py"]],
                    'right_stimuli_normal':  hardlot.loc[t, ["x2", "px2", "y2", "py2"]], 
                    "right_stimuli_big" :    bighlot.loc[t, ["x2", "px2", "y2", "py2"]],
                    'left_feedback_normal':  np.random.choice(a = hardlot.loc[t, ["x", "y"]],  p= hardlot.loc[t, ["px", "py"]]), 
                    "left_feedback_big":     np.random.choice(a = bighlot.loc[t, ["x", "y"]],  p= bighlot.loc[t, ["px", "py"]]), 
                    'right_feedback_normal': np.random.choice(a = hardlot.loc[t, ["x2", "y2"]],  p= hardlot.loc[t, ["px2", "py2"]]), 
                    "right_feedback_big":    np.random.choice(a = bighlot.loc[t, ["x2", "y2"]],  p= bighlot.loc[t, ["px2", "py2"]])}
   
    randomizer = np.random.binomial(n=1, p=.5, size=1)
    if stimuli["stimuli_sequence"] == 0:
        if randomizer == 0 :
            left_stimuli.text = "%s" % stimuli["left_stimuli_normal"]
            left_feedback.text = '%s' % stimuli["left_feedback_normal"]
            right_stimuli.text = "%s" % stimuli["right_stimuli_normal"]
            right_feedback.text = '%s' % stimuli["right_feedback_normal"]
        else:
            left_stimuli.text = "%s" % stimuli["right_stimuli_normal"]
            left_feedback.text = '%s' % stimuli["right_feedback_normal"]
            right_stimuli.text = "%s" % stimuli["left_stimuli_normal"]
            right_feedback.text = '%s' % stimuli["left_feedback_normal"]
    else:
        if randomizer == 0 :
            left_stimuli.text = "%s" % stimuli["left_stimuli_big"]
            left_feedback.text = '%s' % stimuli["left_feedback_big"]
            right_stimuli.text = "%s" % stimuli["right_stimuli_big"]
            right_feedback.text = '%s' % stimuli["right_feedback_big"]
        else:
            left_stimuli.text = "%s" % stimuli["right_stimuli_big"]
            left_feedback.text = '%s' % stimuli["right_feedback_big"]
            right_stimuli.text = "%s" % stimuli["left_stimuli_big"]
            right_feedback.text = '%s' % stimuli["left_feedback_big"]
      
    # first event
    trial_routine.wait_for_time_limit(
        components=[fixation_cross], 
        time_seconds=fixation_duration, 
        label='fixation_cross')
    
    # second event
    #antwort registrieren und speichern
    key, rt = trial_routine.wait_for_keys(
        components=[left_stimuli, right_stimuli],
        valid_keys=choice_keys,
        label='gamble_choice')
    
    data = data.append({'trial': t, "choice": key, "left_feedback_normal" : stimuli["left_feedback_normal"], "left_feedback_big": stimuli["left_feedback_big"],
                        "right_feedback_normal": stimuli["right_feedback_normal"], "right_feedback_big": stimuli["right_feedback_big"]
    }, ignore_index=True)
    
    # third event 
    trial_routine.wait_for_time_limit(
        components=[left_feedback, right_feedback], 
        time_seconds=feedback_duration, 
        label= "feedback")
    
    
       
    #save data to file
    for label in expInfo.keys():
        data[label] = expInfo[label]
    data.to_csv(fileName + '.csv')
    
#cleanup
mywin.close()
core.quit()
```
